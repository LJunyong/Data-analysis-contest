# 안전사고발생시각 분석하기

## 분석할 데이터 (2019년)

```python
import pandas as pd
import matplotlib.pyplot as plt

df_2019 = pd.read_excel("★2019~2023 학교안전사고 데이터_수정.xlsx", sheet_name='2019')
df_kid = df_2019[df_2019['학교급'] == '유치원']
df_elementary = df_2019[df_2019['학교급'] == '초등학교']
df_middle = df_2019[df_2019['학교급'] == '중학교']
df_high = df_2019[df_2019['학교급'] == '고등학교']
df_s = df_2019[df_2019['학교급'] == '특수학교']
```

```python
# 2019년도의 유치원 사고발생시각 데이터 분석
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

accident_count = df_kid['사고시간'].value_counts()
# 그래프로 표현하기
plt.rc('font', family = 'Malgun Gothic')
ax = accident_count.plot(kind = 'bar', color = 'skyblue')
plt.title('2019년 유치원 사고시간 분류')
ax.set_xlabel('사고시간')
ax.set_ylabel('건수')
plt.show()
```

```python
df1['사고형태'].value_counts().plot.bar(rot=0, color='skyblue')
plt.title("2022년 초등학교 사고형태 분류")
plt.xlabel("사고형태")
plt.ylabel("건수")
```

```python
df1['사고형태'].value_counts().plot.bar(rot=0, color='skyblue')
plt.title("2022년 중학교 사고형태 분류")
plt.xlabel("사고형태")
plt.ylabel("건수")
```

```python
df1['사고형태'].value_counts().plot.bar(rot=0, color='skyblue')
plt.title("2022년 고등학교 사고형태 분류")
plt.xlabel("사고형태")
plt.ylabel("건수")
```
